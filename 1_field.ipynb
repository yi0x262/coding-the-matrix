```python
#1.4.1
%matplotlib inline
import matplotlib.pyplot as plt
def complex_plot(list_complex):
    plt.clf()
    for c in list_complex:
        plt.plot(c.real,c.imag,'ob')
    plt.xlabel('real')
    plt.ylabel('imaginary')
    plt.plot()
        
S = {i+2j for i in {2,3}}|{1.75+0.25*i+1j for i in range(7)}
complex_plot(S)
```

```python
#1.4.17
import math
n = 20
w = [math.e**(2*math.pi*i*1j/n) for i in range(n)]
complex_plot(w)
```

```python
class GF2(int):
    def __new__(self,n):
        return super().__new__(self,n!=0)
    def __add__(self,other):
        return self^other
    def __mul__(self,other):
        return self&other

one = GF2(1)
one+one
```
